```python
# Computing Mean and Median for Age and Fare Column
mean_age = X_train["Age"].mean()
median_age = X_train["Age"].median()
mean_fare = X_train["Fare"].mean()
median_fare = X_train["Fare"].median()
```

```python
# Filling in the Null values of the Data with Mean and Median
X_train["Age_Mean"] = X_train["Age"].fillna(mean_age)
X_train["Age_Median"] = X_train["Age"].fillna(median_age)
X_train["Fare_Mean"] = X_train["Fare"].fillna(mean_fare)
X_train["Fare_Median"] = X_train["Fare"].fillna(median_fare)
```

#Missing Numerical Imputation - Using Sklearn

```python
# Splitting the Data
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.3)
```

```python
# Importing SimpleImputer
from sklearn.impute import SimpleImputer
```

```python
# Creating SimpleImputer Object for Mean and Median Imputation
imputer1 = SimpleImputer(strategy="mean")
imputer2 = SimpleImputer(strategy="median")
```

```python
# Importing ColumnTransformer
from sklearn.compose import ColumnTransformer
```

```python
# ColumnTransformer on Age and Fare Column
tnf = ColumnTransformer(transformers=[
    ("median_age", imputer2, ["Age"]),
    ("mean_fare", imputer1, ["Fare"])
], remainder="passthrough")
```

```python
# Fitting and Transforming Training Data Only
X_train_transformed = tnf.fit_transform(X_train)
```

```python
# Transforming Testing Data
X_test_transformed = tnf.transform(X_test)
```

```python
# Parameter for Finding the Mean and Median used for Imputation
tnf.named_transformers_["median_age"].statistics_
```

```python
# Parameter for Finding the Mean and Median used for Imputation
tnf.named_transformers_["mean_fare"].statistics_
```

#Using KNN

```python
# Importing train_test_split
from sklearn.model_selection import train_test_split
```

```python
# Dividing Features and Target Variables
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
# Splitting the Data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Shape of Training and Testing Set
print(X_train.shape, X_test.shape)
```

```python
# Importing KNNImputer
from sklearn.impute import KNNImputer
```

```python
# Creating KNNImputer Object
knn = KNNImputer(n_neighbors=2)
```

```python
# Fit and Transform is called on Training Data only
X_train_knn = knn.fit_transform(X_train)
```

```python
# Transformation on Testing Data
X_test_knn = knn.transform(X_test)
```

```python
# Importing LogisticRegression
from sklearn.linear_model import LogisticRegression
```

```python
# Creating LogisticRegression Object for KNNImputer
lr_knn = LogisticRegression()
```

```python
# Fitting the Data to LogisticRegression
lr_knn.fit(X_train_knn, y_train)
```

```python
# Prediction from the Trained Model
y_pred_knn = lr_knn.predict(X_test_knn)
```

```python
# Importing Accuracy Score
from sklearn.metrics import accuracy_score
```

```python
# Accuracy Score of the KNNImputed Data
accuracy_score(y_test, y_pred_knn)
```
